```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("NotebookConsumer") \
    .getOrCreate()
spark.sparkContext.setLogLevel("WARN")
```

```python
df = spark.readStream \
    .format("kafka") \
    .option("kafka.bootstrap.servers", "kafka:9092") \
    .option("subscribe", "weather_transformed") \
    .load()

from pyspark.sql.functions import from_json, col
from pyspark.sql.types import StructType, DoubleType, StringType, TimestampType

schema = StructType() \
    .add("temperature", DoubleType()) \
    .add("windspeed", DoubleType()) \
    .add("event_time", TimestampType()) \
    .add("wind_alert_level", StringType()) \
    .add("heat_alert_level", StringType())

parsed_df = df.selectExpr("CAST(value AS STRING)") \
    .select(from_json(col("value"), schema).alias("data")) \
    .select("data.*")
```
